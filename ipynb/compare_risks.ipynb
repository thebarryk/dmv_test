Before ip network is added to the risk database it is broken down into a list of contiguous parent and children networks such that none overlap. The children keep their original risks and the parents get the new risk.

This program investigates how many of the children risks are different from their parents.

Was all that effort worth it?

```python
from ipaddress import *
import json
import xmywhois as mywhois

def read_families(fn='families.json'):
    # read the json that was stored in a file
    with open(fn, 'r') as hi:
        j_families = hi.read()
    families = json.loads(j_families)
    return families

net = lambda x: ip_network(x)

def main():

    def progress(n):
        # Progress report generator
        c = 0
        while True:
            c += 1
            ctr = f'{c} of {n}'
            yield f'{ctr:15} {str(original_parent)}'


    families = read_families()

    risk_class = mywhois.Risk('xmywhois')
    risk = risk_class.risk

    # Look over all the families. 

    progress_report = progress(len(families))
    differences = {}

    for original_parent, family in families.items():

#         print(next(progress_report))

        parents = family['parents']

        for parent in parents:
            children = family['children']
            parent_risk = risk[net(parent)]['score']    

            for child in children:
                child_risk = risk[net(child)]['score']
#                 if parent_risk == child_risk:
                if not parent_risk == child_risk:
                    continue
                if parent in differences:
                    differences[parent]['child_risk'].append((child, child_risk))
                else:
                    differences[parent] = {'parent_risk' : parent_risk,
                                           'child_risk'  : [(child, child_risk)]}
    return families, risk, differences

families, risk, differences = main()
    
```
